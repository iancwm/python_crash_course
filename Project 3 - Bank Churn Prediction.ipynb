## 1.1 - Basic Descriptive Statistics

```python
df.rename(columns={'Naive_Bayes_Classifier_Attrition_Flag_Card_Category_Contacts_Count_12_mon_Dependent_count_Education_Level_Months_Inactive_12_mon_1':'nbclassifier_1','Naive_Bayes_Classifier_Attrition_Flag_Card_Category_Contacts_Count_12_mon_Dependent_count_Education_Level_Months_Inactive_12_mon_2':'nbclassifier_2'},inplace=True)
```

```python
df.info(verbose=True)
```

```python
df_describe = df.describe().T
df_describe
```

```python
df_describe['CV']=df_describe['std']/df_describe['mean']
kurtosis=df.kurtosis().rename('Kurtosis')
skew = df.skew().rename('Skewness')
df_describe=pd.concat([df_describe,kurtosis,skew],axis=1)
df_describe
```

```python
measures = ['CV','Kurtosis','Skewness']

fig,axes=plt.subplots(ncols=3)

for index, measure in enumerate(measures):
    ax=axes[index]
    df_describe[measure].sort_values().plot(kind='bar',ax=ax)
#     ax.set_aspect('equal')
    ax.set_title(f'Measure - {measure}')

fig.set_size_inches(15,4)
fig.set_dpi(100)
```

## 1.2 - Univariate Analysis

### 1.2.1 - Categorical

```python
def categorical_analysis(df,column):
    fig,axes=plt.subplots(ncols=2)
    
    sns.countplot(df[column],ax=axes[0])
    axes[0].set_title(f"{column} Countplot")
    axes[0].set_xticklabels(axes[0].get_xticklabels(), rotation = 90)
    
    df[column].value_counts().plot.pie(autopct='%.2f',ax=axes[1])
    axes[1].set_title(f"{column} Pie Chart")
    
    fig.suptitle(f"Categorical Analysis of {column}")
    fig.tight_layout()
    fig.set_size_inches(8,6)
    fig.set_dpi(100)
    plt.show()
```

```python
for column in df.dtypes[df.dtypes=='object'].index:
    categorical_analysis(df,column)
```

### 1.2.2 - Numerical

```python
df.dtypes[df.dtypes!='object'].index
```

```python
def distribution_analysis(df,column):
    fig,ax=plt.subplots()
    sns.histplot(df[column],kde=True,ax=ax)
    ax.axvline(df[column].median(),ls='--')
    fig.suptitle(f"{column} Distribution Analysis")
```

```python
for column in df.dtypes[df.dtypes!='object'].index:
    distribution_analysis(df,column)
```

### 1.2.3 - Discrete Numerical

```python
discrete_numerical=['Dependent_count','Total_Relationship_Count','Months_Inactive_12_mon','Contacts_Count_12_mon']
for column in discrete_numerical:
    categorical_analysis(df,column)
```

## 1.3 - Multivariate Analysis

### 1.3.1 Total_Trans_Amt

```python
def facet_kdeplot(df, col, row, variable):
    g = sns.FacetGrid(df,col=col,row=row,hue=row)
    g.map(sns.kdeplot,variable)
    g.map(lambda x, **kws: plt.axvline(x.median(),ls='--',color='gray'),variable)

    for ax in g.axes.flat:
        ax.set_xticklabels(ax.get_xticklabels(),rotation=90)
    
    g.set_titles(row_template = '{row_name}', col_template = '{col_name}')
    g.figure.suptitle(f"Distribution of {variable} over {col} and {row}",size=24,y=1.05)
    g.figure.supxlabel(col,size=16,x=0.5,y=-0.05)
    g.figure.supylabel(row,size=16,x=-0.05,y=0.5)
```

```python
facet_kdeplot(df, 'Income_Category','Education_Level','Total_Trans_Amt')
```
